```python
import numpy as np
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import minmax_scale
```

```python
# loading dataset

dataset = np.loadtxt("student_data.csv", delimiter=',', skiprows=1)

# Now we have to split the features and output
# first column contains the output(y) and rest of the other columns contains the features

X = dataset[:,1:4]
Y = dataset[:, 0]

# X contains the data of each student on one row. So dimensions are (400, 3)
# Y is a 1D array of length 400
```

```python
# spliting the dataset into 80% for training and 20% for testing

train_x_orig = X[0:321]
test_x_orig = X[321:400]

train_y = Y[:321]
test_y = Y[321:400]

print(train_x.shape)
print(test_x.shape)

print(train_y.shape)
print(test_y.shape)
```

```python
# lets normalize the dataset. Now each feature is in the range of 0 and 1 with the previous dimensions.

train_x = minmax_scale(train_x_orig)
test_x = minmax_scale(test_x_orig)
```

```python
# Building the model using Keras

model = Sequential()
model.add(Dense(5, input_dim = 3, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))

# The most confusing thing here is that the shape of the input to the model is defined as 
# an argument on the first hidden layer. This means that the line of code that adds the first Dense layer 
# is doing 2 things, defining the input or visible layer and the first hidden layer.
```

```python
# compile the keras model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(train_x, train_y, epochs=500, batch_size=30)
```

```python
_, accuracy = model.evaluate(train_x, train_y)
print('Accuracy of Training set: %.2f' % (accuracy*100))

_, accuracy = model.evaluate(test_x, test_y)
print('Accuracy of Testing set: %.2f' % (accuracy*100))
```
